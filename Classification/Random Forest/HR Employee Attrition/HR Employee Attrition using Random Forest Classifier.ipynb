# Import libraries

```python
# load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Dataset

- Dataset taken from: https://www.kaggle.com/code/faressayah/decision-trees-random-forest-for-beginners

```python
dataset = pd.read_csv('HR-Employee-Attrition.csv')
```

```python
dataset
```

# Get Matrix of Features & Dependent Variable Vector

```python
# Observe that though these might have meaningful info, they still have the same value for each record
# so, they are unnecessary columns
print((dataset['Over18'] == 'Y').all())
print((dataset['StandardHours'] == 80).all())
```

```python
# drop unnecessary columns
dataset.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis = 'columns', inplace = True)
```

```python
dataset
```

```python
# get matrix of features and dependent variable vector 
x = dataset.drop(['Attrition'], axis = 'columns') # this is are dataframe, later it them into numpy ndarray
y = dataset['Attrition'].values
```

```python
x
```

```python
print(y)
```

```python
# get all features and categorical features
features = list(x.columns)
categorical_features = [] # store the indices of the columns
for i, feature in enumerate(features):
    if isinstance(x.iloc[0][feature], str):
        categorical_features.append(i)
```

```python
print(len(features))
print(features)
```

```python
print(len(categorical_features))
print(categorical_features)
```

```python
# convert the dataframe into ndarray (matrix of features)
x = x.iloc[:, :].values
print(type(x))
print(x.shape)
print(x)
```

# Encoding Categorical Data

```python
# encode matrix of features
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

column_transformer = ColumnTransformer(
    transformers = [('encoder', OneHotEncoder(), categorical_features)],
    remainder = 'passthrough'
)

x = column_transformer.fit_transform(x)
```

```python
print(x.shape)
print(x[0])
print(x)
```

```python
# encode target variable
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y = label_encoder.fit_transform(y)
```

```python
print(y.shape)
print(y[0])
print(set(y)) # distinct output classes
print(y)
```

# Split the dataset into training & test set

```python
# split dataset into training & test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
print(type(x_train))
print(x_train.shape)
print(x_train)
```

```python
print(type(y_train))
print(y_train.shape)
print(y_train)
```

```python
print(type(x_test))
print(x_test.shape)
print(x_test)
```

```python
print(type(y_test))
print(y_test.shape)
print(y_test)
```

# Training the Random Forest Classifier

- https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

- Below we are training a random forest classifier, with no restriction on the `max_depth`, and it used `100` decision trees (default), and used `gini` as the splitting criterion.

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

# train the model
clf.fit(x_train, y_train)
```

# Predicting Test set results

```python
y_pred = clf.predict(x_test)
print(y_pred)
```

# Computing the Confusion Matrix, Classification Report & Accuracy Score

```python
# confusion matrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

confusion_matrix(y_test, y_pred)
```

```python
# accuracy score
accuracy_score(y_test, y_pred)
```

```python
# classification report
print(classification_report(y_test, y_pred))
```

> **NOTE:** In order to visualize a decision tree from the random forest, follow this link: https://towardsdatascience.com/4-ways-to-visualize-individual-decision-trees-in-a-random-forest-7a9beda1d1b7

# Training a Decision Tree model, to compare results

```python
from sklearn.tree import DecisionTreeClassifier

clf_decision_tree = DecisionTreeClassifier()

# train the decision tree model
clf_decision_tree.fit(x_train, y_train)

# predict test set results
y_pred_decision_tree = clf_decision_tree.predict(x_test)

# view confusion matrix
print(confusion_matrix(y_test, y_pred_decision_tree))

# view accuracy score
print(accuracy_score(y_test, y_pred_decision_tree))

# view classification report
print(classification_report(y_test, y_pred_decision_tree))
```
